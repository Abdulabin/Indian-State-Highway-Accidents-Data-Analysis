### What were the top five Indian states & UTs with the highest number of injured individuals in state highway accidents in 2018?

```python
#  YEAR 2018
plt.figure(figsize=(13,8))
df=idf.sort_values(by="i2018",ascending=False).head(20)
bp=sns.barplot(y="states",x="i2018",data=df)
for bars in bp.containers:
    bp.bar_label(bars)
bp.set(ylabel="States",xlabel=" Injured in Accident",title="Accidents in Year 2018")
plt.xticks(rotation=70)
plt.grid(axis="x",alpha=0.5)
plt.show()
```

## An Comparative Analysis of Data Trends: 2018 vs. 2021

```python
plt.figure(figsize=(19,5))
categories1 = idf.sort_values(by="i2018",ascending=False)["states"].head(15).tail(14)
values1 = idf.sort_values(by="i2018",ascending=False)["i2018"].head(15).tail(14)
values4 = idf.sort_values(by="i2021",ascending=False)["i2021"].head(15).tail(14)
bar_width = 0.20
x1 = np.arange(len(categories1))
x2 = [x + bar_width for x in x1]
plt.bar(x1, values1, width=bar_width, label='2018',color="r",alpha=0.6)
plt.bar(x2, values4, width=bar_width, label='2021',color="g",alpha=0.6)
plt.xticks(x1 + bar_width / 1, categories1 , rotation=45)
plt.text(x=2,y=12000,s="Data suggests a decrease in injuries on state highways in 2021 compared to 2018 in most States & UTs",fontsize=15)
plt.title("Comparision between year 2018 & 2021",fontsize=20)
plt.legend()
plt.grid(axis="y",alpha=0.5)
plt.show()
```

```python
plt.figure(figsize=(19,5))
df=idf.sort_values(by="i2018",ascending=False).head(15).tail(14)
sns.lineplot(y="i2018",x="states",data=df,label="2018",color='r',linewidth=2)
sns.lineplot(y="i2021",x="states",data=df,label="2021",color='b',linewidth=2)
plt.fill_between(df["states"],df["i2018"],df["i2021"],color="g",where=df["i2018"]>df["i2021"],alpha=0.5)
plt.title("Comparision between year 2018 & 2021",fontsize=20)
plt.text(x=2,y=12000,s="Data suggests a decrease in injuries on state highways in 2021 compared to 2018 in most States & UTs",fontsize=15)
plt.legend()
plt.xticks(rotation=40)
plt.grid(axis="y",alpha=0.5)
plt.show()
```

```python
# Adding Percentage column
idf["percent"]=100-idf["i2021"]/idf["i2018"]*100
```

```python
df=idf[["states","percent"]]
df.loc[df['percent'] < -30, 'percent'] = -30
plt.figure(figsize=(19,10))
df1=df.sort_values(by="percent",ascending=False)
bp=sns.barplot(y="states",x="percent",data=df1.dropna())
plt.title("Number of States & UTs in which injuries increased or Decreased (2018 - 2021)",fontsize=20)
plt.text(y=25,x=15,s=">>> Decrease in injuries",fontsize=15,color="g")
plt.text(y=25,x=-25,s="Increase in injuries <<<",fontsize=15,color="r")
plt.grid(axis="x",alpha=0.5)
plt.show()
```

```python
plt.figure(figsize=(3,5))
x=df[df["percent"]>0]["percent"].count()
y=df[df["percent"]<0]["percent"].count()
plt.pie([x,y],colors=["g","r"],labels=["Decrease in Injuries","Increase in Injuries"],autopct="%1.1f%%",pctdistance=0.8)
plt.pie([1],radius=0.55,colors="w")
plt.title("Accident report 2018-2021")
plt.show()
```

```python
print(f"There are {x} out of 31 States & UTs in which injuries has decreased \nThere are {y}  out of 31 States & UTs in which injuries has Increased")
```

#### An Comparative Analysis of Data Trends: 2018 vs. 2021
In our analysis of 36 states and Union Territories (UTs), we found that in the majority of regions (26 out of 31, or 81.2 %), the rate of injuries decreased. These regions include Andaman and Nicobar, Manipur, Sikkim, Haryana, Telangana and etc .

   Conversely, in a minority of regions (6 out of 31, or 18.8 %), the rate of injuries increased. These regions are Dadra and Nagar Haveli, Bihar, Meghalaya and etc. 
  
  Note that data for 5 states and UTs were null during the analysis.

```python
plt.figure(figsize=(19,5))
categories1 = idf.sort_values(by="i2020",ascending=False)["states"].head(15).tail(10)
values1 = idf.sort_values(by="i2020",ascending=False)["i2020"].head(15).tail(10)
values4 = idf.sort_values(by="i2021",ascending=False)["i2021"].head(15).tail(10)
bar_width = 0.20
x1 = np.arange(len(categories1))
x2 = [x + bar_width for x in x1]
plt.bar(x1, values1, width=bar_width, label='2020',color="r",alpha=0.6)
plt.bar(x2, values4, width=bar_width, label='2021',color="g",alpha=0.6)
plt.xticks(x1 + bar_width / 1, categories1 , rotation=45)
plt.text(x=1.5,y=4500,s="Data suggests a increase in injuries on state highways in 2021 compared to 2020 in most States & UTs",fontsize=15)
plt.title("Comparision between year 2018 & 2021",fontsize=20)
plt.legend()
plt.grid(axis="y",alpha=0.5)
plt.show()
```

### Categorize the data into groups (2018 - 2021)

```python
idf.loc[idf["i2018"]>idf["i2021"],"Catageroy"]="Decreased"
idf.loc[idf["i2018"]<idf["i2021"],"Catageroy"]="Increased"
idf.loc[(idf["i2018"]>idf["i2019"]) & (idf["i2019"]>idf["i2020"]) & (idf["i2020"]>idf["i2021"]),
"Catageroy"]="Continues_Decrease"
idf.loc[df["percent"].isnull(),"Catageroy"]="No_record"
idf.loc[(idf["i2018"]<idf["i2019"]) & (idf["i2018"]<idf["i2020"]) & (idf["i2020"]<idf["i2021"]),
"Catageroy"]="Continues_Increase"
gidf=idf.groupby("Catageroy")[["states"]]
gidf.count()
```

```python
gidfx=idf[idf["Catageroy"] != "No_record"]
gidf1=gidfx.groupby("Catageroy",as_index=False)["states"].count()
plt.pie(gidf1["states"],labels=gidf1["Catageroy"],autopct="%1.1f%%",pctdistance=0.8,startangle=40)
plt.pie([1],radius=0.6,colors="w")
plt.title("Percent Catagery")
plt.show()
```

```python
gidf=idf.groupby("Catageroy")[["states"]]
print("Gradually Increase")
print(gidf.get_group("Continues_Increase"))
print("\nGradually Decrease")
print(gidf.get_group("Continues_Decrease"))
```
