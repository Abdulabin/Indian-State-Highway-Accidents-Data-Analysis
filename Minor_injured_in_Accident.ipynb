### States & UTs which are having Highest & Lowest Serious Injuries 

```python
# States & UTs where Number of peoples Seriously Injured in State Highway Accidents are maximum.
# Top 5
plt.figure(figsize=(10,8))
plt.subplot(2,1,1)
df=mdf.sort_values(by="serious_inj",ascending=False).head()
bp=sns.barplot(y="states",x="serious_inj",data=df,palette="summer")
for bars in bp.containers:
    bp.bar_label(bars)
plt.title("Seriously Injured in Accidents ")
plt.ylabel("Top 5 States & UTs")
plt.xticks(rotation=70)
plt.grid(axis="x",alpha=0.5)

# Bottom 5
plt.subplot(2,1,2)
df=mdf[["states","serious_inj"]][mdf["serious_inj"]>0]
df=df.sort_values(by="serious_inj",ascending=False).tail()
bp=sns.barplot(y="states",x="serious_inj",data=df,palette="summer")
for bars in bp.containers:
    bp.bar_label(bars)
bp.set(xlabel="No.of peoples Seriously Injured in Accident")
plt.ylabel("Bottom 5 States & UTs")
plt.xticks(rotation=70)
plt.grid(axis="x",alpha=0.5)

plt.show()
```

### How much percent there are minor & serious injuries in total injuries

```python
# Sum of Injured
plt.figure(figsize=(13,4))
# Bar Chart
plt.subplot(1,2,1)
y=[mdf["minor_inj"].sum(),mdf["serious_inj"].sum()]
x=["minor_inj","serious_inj"]
bp=sns.barplot(y=x,x=y,color="c",saturation=0.4)
for bars in bp.containers:
    bp.bar_label(bars)
bp.set(title="Sum of Injuries")
plt.xlim(40000,50000)
plt.grid(axis="x",alpha=0.5)

# Pie Chart
plt.subplot(1,2,2)
plt.pie(y,labels=["Minor","Serious"],autopct="%1.1f%%",pctdistance=0.8,startangle=-1,colors=["g","r"])
plt.pie([1],colors="w",radius=0.6)
plt.title("Percent of Minor & Serious Injured in Total Injuries")
plt.show()
```

#### How Much States & UTs are having higher serious injuries ratio than Minor Injuries 

```python
# Categorizing Road Injuries as Serious and Not Serious
mdf.loc[mdf["serious_inj"]>mdf["minor_inj"],"Catagorey"]="Serious"
mdf.loc[(mdf["serious_inj"]<mdf["minor_inj"]) | (mdf["total_inj"]==0) ,"Catagorey"]="Not Serious"
gmdf=mdf.groupby("Catagorey")[["states"]]
gmdf.count()
```

```python
gmdf.get_group("Serious").head()
```

#### Top 5 States and UTs where there is more probability of getting Serious Injuries in Accident

```python
# A Comparison of the Rates of Serious and Minor Injuries
# Top 5 states & UTs which are having maximum rate of serious injuries
serdf=mdf.loc[mdf["Catagorey"]=="Serious"]
serdf.loc[:, "%_Serious_Inj"] = serdf["serious_inj"] / serdf["total_inj"] * 100
serdf.sort_values(by="%_Serious_Inj",ascending=False).head()
```

#### Top 5 States and UTs where there is more probability of getting Minor Injuries in Accident

```python
# A Comparison of the Rates of Serious and Minor Injuries
# Top 5 states & UTs which are having maximum rate of Minor injuries
serdf=mdf.loc[mdf["Catagorey"]=="Not Serious"]
serdf["%_Minor_Inj"]=serdf["serious_inj"]/serdf["total_inj"]*100
serdf.sort_values(by="%_Minor_Inj",ascending=False).head()
```
